```python
import cv2
import matplotlib.pyplot as plt
```

# Original image

```python
img_path = "Dataset/img_align_celeba/img_align_celeba/000001.jpg"

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Degraded image

```python
scale = 4
h, w = img.shape[:2]

lr = cv2.resize(img, (w//scale, h//scale), interpolation=cv2.INTER_CUBIC)
```

# Upscale image

```python
lr_up = cv2.resize(lr, (w, h), interpolation=cv2.INTER_CUBIC)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(img)
axs[0].set_title("High Resolution (Original)")
axs[0].axis("off")

axs[1].imshow(lr)
axs[1].set_title("Low Resolution")
axs[1].axis("off")

axs[2].imshow(lr_up)
axs[2].set_title("Upscaled (bicubic)")
axs[2].axis("off")

plt.show()
```
